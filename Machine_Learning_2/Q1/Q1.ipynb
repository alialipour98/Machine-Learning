```python
# Machine Learning
import numpy                    as np
import pandas                   as pd
import matplotlib.pyplot        as plt

from mlxtend.evaluate           import bias_variance_decomp
from sklearn.model_selection    import train_test_split
from sklearn.preprocessing      import PolynomialFeatures
from sklearn.metrics            import mean_squared_error
from sklearn.linear_model       import LinearRegression
from sklearn.pipeline           import make_pipeline
```

```python
# Machine Learning
def make_data(X):
    return (2 * np.cos(X) / np.pi + (2 * X) / (2 * np.pi) + 2 * np.cos(3*X) / (-3 * np.pi)).ravel()
def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree), LinearRegression(**kwargs))
```

```python
# Machine Learning
mean = 0
std  = 1

X = np.arange(-10, 10, 0.1).reshape(-1, 1)
y = make_data(X=X)

y_guassion = y + 0.1*np.random.normal(mean, std, size = y.shape)
y_poission = y + 0.1*np.random.poisson(lam = 2, size = y.shape)
```

```python
# Machine Learning
model1_G        = PolynomialRegression(1 ).fit(X, y_guassion)
model3_G        = PolynomialRegression(3 ).fit(X, y_guassion)
model8_G        = PolynomialRegression(8 ).fit(X, y_guassion)
model15_G       = PolynomialRegression(15).fit(X, y_guassion)

model1_P        = PolynomialRegression(1 ).fit(X, y_poission)
model3_P        = PolynomialRegression(3 ).fit(X, y_poission)
model8_P        = PolynomialRegression(8 ).fit(X, y_poission)
model15_P       = PolynomialRegression(15).fit(X, y_poission)

y1_predict_G    = model1_G.predict (X)
y3_predict_G    = model3_G.predict (X)
y8_predict_G    = model8_G.predict (X)
y15_predict_G   = model15_G.predict(X)

y1_predict_P    = model1_P.predict (X)
y3_predict_P    = model3_P.predict (X)
y8_predict_P    = model8_P.predict (X)
y15_predict_P   = model15_P.predict(X)

MSE_1_P         = mean_squared_error(y1_predict_P , y)
MSE_3_P         = mean_squared_error(y3_predict_P , y)
MSE_8_P         = mean_squared_error(y8_predict_P , y)
MSE_15_P        = mean_squared_error(y15_predict_P, y)

MSE_1_G         = mean_squared_error(y1_predict_G , y)
MSE_3_G         = mean_squared_error(y3_predict_G , y)
MSE_8_G         = mean_squared_error(y8_predict_G , y)
MSE_15_G        = mean_squared_error(y15_predict_G, y)

xfit            = np.linspace(-10, 10, 1000).reshape(-1, 1)
fig, ax         = plt.subplots(nrows = 2, ncols = 4, figsize=(25, 10))

fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.2, hspace = 0.1)

ax[0][0].scatter(X, y_guassion)
ax[0][0].plot(xfit, model1_G.predict(xfit) , color = 'r', label = 'Degree = 1' , linewidth = 4)
ax[0][0].legend()
ax[0][0].get_xaxis().set_ticks([])
ax[0][0].set_title(f'Guassion , Mean Square Error => {round(MSE_1_G, 4)}')
ax[0][0].axis('equal')

ax[0][1].scatter(X, y_guassion)
ax[0][1].plot(xfit, model3_G.predict(xfit) , color = 'r', label = 'Degree = 3' , linewidth = 4)
ax[0][1].legend()
ax[0][1].get_xaxis().set_ticks([])
ax[0][1].set_title(f'Guassion , Mean Square Error => {round(MSE_3_G, 4)}')

ax[1][0].scatter(X, y_guassion)
ax[1][0].plot(xfit, model8_G.predict(xfit) , color = 'r', label = 'Degree = 8' , linewidth = 4)
ax[1][0].legend()
ax[1][0].get_xaxis().set_ticks([])
ax[1][0].set_title(f'Guassion , Mean Square Error => {round(MSE_8_G, 4)}')

ax[1][1].scatter(X, y_guassion)
ax[1][1].plot(xfit, model15_G.predict(xfit), color = 'r', label = 'Degree = 15', linewidth = 4)
ax[1][1].legend()
ax[1][1].get_xaxis().set_ticks([])
ax[1][1].set_title(f'Guassion , Mean Square Error => {round(MSE_15_G, 4)}')

ax[0][2].scatter(X, y_poission)
ax[0][2].plot(xfit, model1_P.predict(xfit) , color = 'r', label = 'Degree = 1' , linewidth = 4)
ax[0][2].legend()
ax[0][2].get_xaxis().set_ticks([])
ax[0][2].set_title(f'Poisson , Mean Square Error => {round(MSE_1_P, 4)}')

ax[0][3].scatter(X, y_poission)
ax[0][3].plot(xfit, model3_P.predict(xfit) , color = 'r', label = 'Degree = 3' , linewidth = 4)
ax[0][3].legend()
ax[0][3].get_xaxis().set_ticks([])
ax[0][3].set_title(f'Poisson , Mean Square Error => {round(MSE_3_P, 4)}')

ax[1][2].scatter(X, y_poission)
ax[1][2].plot(xfit, model8_P.predict(xfit) , color = 'r', label = 'Degree = 8' , linewidth = 4)
ax[1][2].legend()
ax[1][2].get_xaxis().set_ticks([])
ax[1][2].set_title(f'Poisson , Mean Square Error => {round(MSE_8_P, 4)}')

ax[1][3].scatter(X, y_poission)
ax[1][3].plot(xfit, model15_P.predict(xfit), color = 'r', label = 'Degree = 15', linewidth = 4)
ax[1][3].legend()
ax[1][3].get_xaxis().set_ticks([])
ax[1][3].set_title(f'Poisson , Mean Square Error => {round(MSE_15_P, 4)}')

plt.ylim([-0,   6])
plt.ylim([-10, 10])
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
# estimate bias and variance
mse, bias, var = bias_variance_decomp(model1_P, X_train, y_train, X_test, y_test, loss='mse', num_rounds=200, random_seed=1)
# summarize results
print('Noise : Poisson (Degree >> 1)')
print('MSE\t\t\t\t: %.3f' % mse)
print('Bias\t\t\t: %.3f' % bias)
print('Variance\t: %.3f' % var)
print('------------------------------')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
# estimate bias and variance
mse, bias, var = bias_variance_decomp(model3_P, X_train, y_train, X_test, y_test, loss='mse', num_rounds=200, random_seed=1)
# summarize results
print('Noise : Poisson (Degree >> 3)')
print('MSE\t\t\t\t: %.3f' % mse)
print('Bias\t\t\t: %.3f' % bias)
print('Variance\t: %.3f' % var)
print('------------------------------')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
# estimate bias and variance
mse, bias, var = bias_variance_decomp(model8_P, X_train, y_train, X_test, y_test, loss='mse', num_rounds=200, random_seed=1)
# summarize results
print('Noise : Poisson (Degree >> 8)')
print('MSE\t\t\t\t: %.3f' % mse)
print('Bias\t\t\t: %.3f' % bias)
print('Variance\t: %.3f' % var)
print('------------------------------')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
# estimate bias and variance
mse, bias, var = bias_variance_decomp(model15_P, X_train, y_train, X_test, y_test, loss='mse', num_rounds=200, random_seed=1)
# summarize results
print('Noise : Poisson (Degree >> 15)')
print('MSE\t\t\t\t: %.3f' % mse)
print('Bias\t\t\t: %.3f' % bias)
print('Variance\t: %.3f' % var)
```
