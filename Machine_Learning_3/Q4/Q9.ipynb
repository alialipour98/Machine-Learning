```python
import pandas                   as pd
import matplotlib.pyplot        as plt

from sklearn.model_selection    import train_test_split
from sklearn.tree               import DecisionTreeClassifier
from sklearn.metrics            import accuracy_score
from sklearn.tree               import plot_tree
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
dfd = df.iloc[:, 0:-1]
t   = df.iloc[:, [-1]]
```

```python
dfd.head();
```

```python
t.head();
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(dfd, t, test_size = 0.3, random_state = 1)

X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 1)

tree.fit(X_train, Y_train);
```

```python
yPred_Train = tree.predict(X_train)
yPred_Test  = tree.predict(X_test)

acc_Train   = accuracy_score(Y_train, yPred_Train)
acc_Test    = accuracy_score(Y_test, yPred_Test)

print(f'Accuracy For Train Data : {acc_Train}')
print(f'Accuracy For Test  Data : {acc_Test} ')
```

```python
fig, ax = plt.subplots(figsize=(20, 10))  # whatever size you want
plot_tree(tree, max_depth = 3, impurity = True, fontsize = 10, ax = ax);
```

```python
tree = DecisionTreeClassifier(criterion = 'entropy', max_depth = 2, random_state = 1)

tree.fit(X_train, Y_train);
```

```python
yPred_Train = tree.predict(X_train)
yPred_Test  = tree.predict(X_test)

acc_Train   = accuracy_score(Y_train, yPred_Train)
acc_Test    = accuracy_score(Y_test, yPred_Test)

print(f'Accuracy For Train Data : {acc_Train}')
print(f'Accuracy For Test  Data : {acc_Test} ')
```
